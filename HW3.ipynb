### Вывести на экран первые 5 строк

* 1.1 Создать новый признак price_per_sq_lot, который будет содержать среднюю стоимость за один кв. метр общей площади

* 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
* Если реновации дома не было, то в новом признаке поставьте 0

```python
df.head(5)
```

```python
df.info()
```

```python
def price_per_sq(a,b):
    return a / b
```

```python
df['price_per_sq_lot'] = df.apply(lambda x: price_per_sq(x['price'], x['sqft_lot']), axis = 1)
```

```python
df.head(5)
```

```python
def delta(a,b):
    return a - b
```

```python
df['delta_renovated'] = df.apply(lambda x: delta(x['yr_renovated'], x['yr_built']), axis = 1)
```

```python
df['delta_renovated'] = df['delta_renovated'].apply(lambda x: x if x > 0 else 0)
```

```python
df.head(5)
```

```python
df['year_sold'] = 0
df['month_sold'] = 0
```

```python
df.head(5)
```

```python
df.drop(['date','zipcode','lat','long'], axis=1, inplace=True)
df.head(5)
```

```python
def gold_max(x):
    m = x.to_numpy().max() 
    color = {True: "background-color: #c78f2e", False: ""}
    is_max = (x == m).replace(color)
    return is_max
```
